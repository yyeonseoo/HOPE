```python
import fitz

pdf_path = "2015개정경제수학-광주교육청.pdf"

doc = fitz.open(pdf_path)

print("PDF 파일명:", pdf_path)
print("전체 페이지 수:", len(doc))
```

```python
sample_pages = [0, 3, 6, 7, 10, 14, 15, 16]

for page_num in sample_pages:
    page = doc[page_num]
    text = page.get_text().strip()

    print(f"{page_num + 1}페이지 텍스트 길이:", len(text))
    print(text[:200])
    print("-" * 50)
```

해당 PDF는 총 198페이지로 구성되어 있음을 확인하였다.

또한 표지, 본문, 수식, 표, 그래프가 포함된 샘플 페이지를 대상으로
